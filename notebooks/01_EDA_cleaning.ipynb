## 2. Data Cleaning & Preprocessing

```python
VALID_STATUTS   = ["Terminée"]
RETURN_STATUTS  = ["Retournée", "Remboursée"]

df_returns = df[df["statut_commande"].isin(RETURN_STATUTS)].copy()
df = df[df["statut_commande"].isin(VALID_STATUTS)].copy()

print(f"✓ Terminée: {len(df):,}  |  Returns/refunds kept aside: {len(df_returns):,}")
print(f"  Dropped (other statuts): {17263 - len(df) - len(df_returns):,}")
```

```python
df["date_commande"] = pd.to_datetime(df["date_commande"], errors="coerce")
df["quantite"]      = df["quantite"].astype("Int64")   # nullable int
df["montant_da"]    = df["montant_da"].astype(float)
before = len(df)
```

```python

df = df.dropna(subset=["date_commande"])
df = df.dropna(subset=["quantite", "montant_da"])
print(f"✓ Dropped {before - len(df)} rows (missing date or revenue)")
df["produit"] = df["produit"].fillna("Inconnu")
```
